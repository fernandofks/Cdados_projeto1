## Pergunta (s) a ser respondida

A OECE(Organização Europeia de Cooperação Econômica) foi um organismo que surgiu a partir do plano Marshall dos Estados Unidos. Assim, surge a pergunta: Como foram os resultados econômicos da OECE nos países participantes?

Link das informações: http://www.oecd.org/about/budget/member-countries-budget-contributions.htm

## Explicação da hipótese e mecanismo

**Hipótese:** Participar da OECE logo após(em 1948) a Segunda Guerra Mundial garantia a melhora da situação economica dos países em comparação com os que não participaram.

(Inicialmente participaram Alemanha, Austria, Bélgica, Dinamarca, França, Reino Unido, Itália, Noruega, Portugal, Suécia, Suíça e Turquia)

(Dos participantes na Segunda Guerra Mundial, não participaram da OEC: Romania, Bulgaria, Húngria, Rússia, Polanda, Luxemburgo e Grécia)

**Mecanismo:** A OECE tinha como um dos objetivos dividir os fundos recebidos pelos Estados Unidos, assim, pensa-se que por receberem fundos dos Estados Unidos, sua situação econômica melhoraria mais do que a dos demais países que participaram da guerra.

## Segunda pergunta a ser respondida

Pela conclusão anterior, é possível concluir que a OECE não garantiu uma estabilização ou crescimento do GDP melhor que os países que participaram da Segunda Guerra mas não entraram na OECE. Contudo, fica uma questão: A ajuda recebida pelo plano Marshall ajudou o GDP dos países da OECE? Talvez, até uma outra pergunta seja necessária: A ajuda econômica que os países recebem aumentam seu GDP? 

## Explicação da hipótese e mecanismo

**Hipótese:** Existe uma grande correlação entre o quanto de ajuda econômica (aid received) que um país recebe e o crescimento de seu GDP.

**Mecanismo:** Os países que recebem ajuda de outros conseguem manusear essa ajuda econômica de modo a aumentar seu GDP consequentemente.

## Bases de dados usadas

Além do GDP conseguido pelo Gapminder, também foi necessário os dados de Aid Received do Gapminder.

## Análise de dados

```python
#lendo os dados e setando o index

aid = pd.read_excel('aid_received.xlsx')
aid_index=aid.set_index("country")
```

```python
#Fazendo join da tabela(Aprendizado importante: Fazer join com o index nos países, e não nos anos)

dados_join = dados_index.join(aid_index, how='inner', lsuffix='_GDP', rsuffix='_aid').T
dados_join.tail()
```

```python
#Relação de número de países na base de dados que participaram e não participaram da OECE
#--Explicação do porquê esse gráfico é importante a seguir.
for i in dados_t.columns:
    dados_t.loc["OECE",i]="nonOECE"
for j in ["Germany","Austria","Belgium","Denmark","France","United Kingdom","Italy","Norway","Portugal","Switzerland","Sweden","Turkey"]:
    dados_t.loc["OECE",j]="OECE"
a = dados_t.T.groupby('OECE').size()
plt.pie(a, labels=["Participaram da OECE","Não participaram da OECE"])
plt.show()
```

```python
#Usando os dados de um ano aleatório(1960) para ver se existe alguma relação entre aid received e crescimento do GDP
#Não foram usados os anos da Guerra pois não existiam os dados nesse período e não foram usados os países 
#participantes pois quero ver um conceito geral, isto é, se a ajuda financeira a um pais aumenta seu GDP
#consequentemente. Assim, vale ressaltar que o número de países que participaram da OECE é muito menor do que os 
#que não participaram, então seria interessante validar esse conceito(hipótese) com todos os países.
dados_1960_aid=dados_join.loc["1960_aid",:]
dados_1960_GDP=dados_join.loc["1960_GDP",:]
plt.figure(figsize=(15,10))
plt.scatter(dados_1960_aid,dados_1960_GDP)
plt.title("Relação entre Aid Received e Crescimento do GDP")
plt.xlabel("Aid Received em Dólares")
plt.ylabel("Crescimento do GDP per Capita")
plt.xscale("log")
plt.xlim(1e5, 1e10)
plt.plot()
print("A correlação é de {0}".format(dados_1960_aid.corr(dados_1960_GDP)))
```

## Conclusão

**Resultados:** A correlação entra a ajuda recebida e o crescimento do GDP dos países não está nada relacionada, uma vez que quase não há correlação entre os dois dados.

**Conclusão:** Há uma grande probabilidade de que a ajuda distribuída pela OECE aos países participantes não ajudou no crescimento econômico deles, uma vez que o primeiro resultado mostra que os países que não receberam a ajuda do plano Marshall tiveram um similar crescimento de GDP e que o segundo resultado mostra que a ajuda financeira a países não está intimamente ligada ao crescimento de seu GDP.
